**Вопрос 1**: Опишите, как изменяется адаптивная способность алгоритма ЭСС при изменении параметра $\alpha$ от 0 до 1.

**Ответ**: Алгоритм ЭСС: $\widehat{y}_{t+1}=\widehat{y}_{t} + \alpha(y_{t}-\widehat{y}_{t})$. Чем ближе $\alpha$ к 0, тем больше последних наблюдений учитывается (приблизительно $1/\alpha$). При $\alpha=1$ прогноз -- последнее наблюдение, адаптивная способность больше; при $\alpha\rightarrow 0$ получается что-то вроде среднего по всем наблюдениям, адаптивная способность меньше.

**Вопрос 2**: Каким следует выбрать параметр сглаживания тренда $\alpha_2$ в модели Хольта (линейный тренд) в случае, когда вы предсказываете временной ряд 1) с плавно меняющимя трендом; 2) стохастически меняющися трендом?

**Ответ**: Модель Хольта:
\begin{align*}
&\widehat{y}_{t+d} = a_{t} + db_{t} \\
&a_{t} = \alpha_{1}y_{t} + (1 - \alpha_{1})(a_{t-1} - b_{t-1}) \\
&b_{t} = \alpha_{2}(a_{t} - a_{t-1}) + (1 - \alpha_{2})b_{t-1}
\end{align*}
При малых $\alpha_2$ модель сильнее усредняет тренд и больше подходит дла случая 2); если же тренд меняется плавно, то можно помнить лишь информацию о недалеком прошлом, поэтому подойдут $\alpha_2$ близкие к 1.

**Вопрос 3**: Выпишите формулу временного ряда, соответствующих следующим ARIMA моделям
  
  1) ARIMA $(p=0,q=1,d=s)$
  
  2) ARIMA $(p=0,q=s+1,d=1)(P=1, Q=0, D=1)$

**Ответ**:

1) $\Delta^{s}y_{t} = \alpha_{t} + \varepsilon_{t} + \theta_{1}\varepsilon_{t-1}$.

2) $\Delta^{1}\Delta_{S}^{1}y_{t} = \alpha_{t} + \varepsilon_{t} + \theta_{1}\varepsilon_{t-1} + \ldots + \theta_{s+1}\varepsilon_{t-s-1} + \varphi_{S}y_{t-S}$.

Где $\Delta_{S}^{D}$ -- cезонное дифференцирование, примененное $D$ раз, $\Delta_{S}$ -- обычное дифференцирование, примененное $d$ раз, $S$ -- период сезонности.

**Вопрос 4**:  Приведите пример такой модели типа ЭС (экспоненциальное сглаживание), которой нет эквивалента среди семейства ARIMA. 

**Ответ**: Судя по определениям с лекции ЭС = ЭСС, а ЭСС = $ARMA(1,1)$, так что такой модели нет.

**Вопрос 5**:  В чем принципиальное отличие между адаптивной композицей и адаптивной селективной моделью прогнозирования?
В какие случаях предполчительнее использовать адаптивную селекцию нежели адаптивную композцию?

**Ответ**: Адаптивная селекция выбирает одну модель, которая лучше всего предсказывает временной ряд в данный момент времени. Адаптивная комозиция строит композицию всех моделей. Адаптивная селекция предпочтительнее в случае, когда можели сильно отличаются по качеству и мы можем быстро и эффективно выбрать наилучшую.

**Вопрос 6**: Что такое hold-out период и out-of-sample период?

**Ответ**: hold-out period = validation period, период для подбора гиперпараметров; out-of-sample period = test period, пероид для оценки модели.

** Вопрос 7**: Опишите, что такое автокорреляционная функция и частная автокорреляционная функция.
Зачем они нужны?

**Ответ**: Автокорреляционная функция:
$$
r_{\tau}=corr(y_{t}, y_{t+\tau}) = \frac{\mathsf{E}((y_t-\mathsf{E}y)(t_{t+\tau}-\mathsf{E}y))}{\mathsf{D}y}.
$$

Частичная автокорреляционная функция:
$$
part\_r_{\tau}=corr(y_{t} - AR^{\tau-1}_{t}, y_{t+\tau} - AR^{\tau-1}_{t+\tau}),
$$
где $AR_{t}^{p}$ -- авторегрессия порядка $p$.

По коррелограммам (графикам автокорреляций) можно выяснять наличие в ряде разных компонент: тренда, сезонности, циклов.

```python
def WintersExponentialSmoothing(x, h, Params):
    T = len(x)
    alpha = Params['alpha']
    delta = Params['delta']
    s = Params['seasonality_period']
    
    FORECAST = np.zeros(T + h)
    
    a = 0
    b = np.zeros(T + s)
    
    for t in range(T):
        if t < s:
            if np.isnan(x[t]):
                continue
            b[t] = delta * (x[t] - a)
            continue
            
        if not np.isnan(x[t]):
            a = alpha * (x[t] - b[t - s]) + (1 - alpha) * a
            b[t] = delta * (x[t] - a) + (1 - delta) * b[t - s]
        FORECAST[t + h] = a + b[t - s + h % s]
    return FORECAST
```
